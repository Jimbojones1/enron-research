```python
import pandas as pd
import numpy as np
# this module allows you to parse emails
import email
```

```python
emails_df = pd.read_csv('emails.csv')
```

```python
emails_df.head()
```

```python
## This code is creating a list object (an array) from a python data series
## (emails_df,) and were using map to parse the email into an array objects
## using the email module function (email.message_from_string)
messages = list(map(email.message_from_string, emails_df['message']))

# drop the email message from the dateframe since we don't need it anymore
emails_df.drop('message', axis=1, inplace=True)
```

```python
print(messages[0]['X-From'])
```

```python
### Get all the fields from the emails
keys = messages[0].keys()
```

```python
print(keys)
```

```python
### this code looping through all the messages which is  a list of objects
## and assigin that to the value for each email
for key in keys:
    emails_df[key] = [doc[key] for doc in messages]
```

```python
emails_df.head()
```

```python
def get_text_from_email(msg):
    '''To get the content from email objects'''
    parts = []
   
    ## The walk() method is an all-purpose geneartor which can be used to 
    ## iterate over all the parts and subparts or a message object tree
    ## typically used as an iterator in for loop
    for part in msg.walk():
        
        if part.get_content_type() == 'text/plain':
            parts.append( part.get_payload() )
  
    return ''.join(parts)
```

```python
# Parse content from emails
emails_df['content'] = list(map(get_text_from_email, messages))


```
